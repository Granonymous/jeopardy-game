## 1. Importing Modules

Python has tons of built-in modules (the "standard library"). You import them to use their functionality.

````python
# Import entire module
import math
print(math.sqrt(16))
print(math.pi)
````

````python
# Import specific items
from math import sqrt, pi
print(sqrt(16))
print(pi)
````

````python
# Import with alias
import math as m
print(m.sqrt(16))
````

## 2. Standard Library Highlights

Python's standard library has modules for almost everything!

````python
# random - random numbers
import random
print(f"Random int 1-10: {random.randint(1, 10)}")
print(f"Random choice: {random.choice(['a', 'b', 'c'])}")
print(f"Shuffled: {random.sample([1, 2, 3, 4, 5], 5)}")
````

````python
# datetime - dates and times
from datetime import datetime, date
print(f"Now: {datetime.now()}")
print(f"Today: {date.today()}")
print(f"Year: {date.today().year}")
````

````python
# collections - useful data structures
from collections import Counter
words = ["apple", "banana", "apple", "cherry", "apple", "banana"]
word_counts = Counter(words)
print(word_counts)
print(f"Most common: {word_counts.most_common(2)}")
````

````python
# pathlib - file paths (modern way!)
from pathlib import Path

# Get current directory
here = Path(".")
print(f"Current dir: {here.absolute()}")

# Build paths (no more os.path.join!)
data_path = Path("data") / "sample.json"
print(f"Data path: {data_path}")
````

## 3. Creating Your Own Modules

Any `.py` file is a module! If you create `mymodule.py`, you can import it with `import mymodule`.

```python
# mymodule.py
def greet(name):
    return f"Hello, {name}!"

PI = 3.14159
```

```python
# main.py
import mymodule
print(mymodule.greet("Alice"))
print(mymodule.PI)
```

## 4. The `__name__ == "__main__"` Pattern

This is a super common pattern! It lets code run when a file is executed directly, but NOT when imported.

````python
# In a module, __name__ tells you how the file was loaded
print(f"In this notebook, __name__ is: {__name__}")
````

When you run a file directly: `__name__` is `"__main__"`  
When you import a file: `__name__` is the module name

```python
# calculator.py
def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

# This only runs if calculator.py is executed directly
if __name__ == "__main__":
    print("Testing calculator:")
    print(f"2 + 3 = {add(2, 3)}")
    print(f"5 - 2 = {subtract(5, 2)}")
```
