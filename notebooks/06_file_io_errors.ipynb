## 1. Reading Files

````python
# Always use 'with' - it automatically closes the file
with open("../README.md", "r") as f:
    content = f.read()

print(content[:200])  # First 200 characters
````

````python
# Read line by line
with open("../README.md", "r") as f:
    for i, line in enumerate(f):
        print(f"{i}: {line.strip()}")
        if i >= 5:
            break
````

## 2. Writing Files

````python
# Write mode - creates or overwrites
with open("data/test_output.txt", "w") as f:
    f.write("Hello, World!\n")
    f.write("This is line 2.\n")

# Verify it worked
with open("data/test_output.txt", "r") as f:
    print(f.read())
````

````python
# Append mode - adds to existing file
with open("data/test_output.txt", "a") as f:
    f.write("This is line 3 (appended).\n")

with open("data/test_output.txt", "r") as f:
    print(f.read())
````

## 3. Pathlib (Modern Path Handling)

````python
from pathlib import Path

# Create a path
data_dir = Path("data")
file_path = data_dir / "sample.txt"  # Use / to join paths!

print(f"Path: {file_path}")
print(f"Exists: {file_path.exists()}")
print(f"Parent: {file_path.parent}")
print(f"Name: {file_path.name}")
print(f"Suffix: {file_path.suffix}")
````

````python
# List files in a directory
notebooks_dir = Path(".")
for f in notebooks_dir.glob("*.ipynb"):
    print(f.name)
````

## 4. JSON Files

````python
import json

# Python dict to JSON string
data = {
    "name": "Alice",
    "age": 25,
    "hobbies": ["reading", "coding"]
}

json_string = json.dumps(data, indent=2)
print(json_string)
````

````python
# Write JSON to file
with open("data/person.json", "w") as f:
    json.dump(data, f, indent=2)

# Read JSON from file
with open("data/person.json", "r") as f:
    loaded = json.load(f)

print(loaded)
print(loaded["name"])
````

## 5. CSV Files

````python
import csv

# Write CSV
students = [
    {"name": "Alice", "grade": 95},
    {"name": "Bob", "grade": 87},
    {"name": "Charlie", "grade": 92}
]

with open("data/students.csv", "w", newline="") as f:
    writer = csv.DictWriter(f, fieldnames=["name", "grade"])
    writer.writeheader()
    writer.writerows(students)
````

````python
# Read CSV
with open("data/students.csv", "r") as f:
    reader = csv.DictReader(f)
    for row in reader:
        print(f"{row['name']}: {row['grade']}")
````

## 6. Error Handling: try/except

````python
# Without error handling - crashes!
# x = int("not a number")
````

````python
# With error handling
try:
    x = int("not a number")
except ValueError:
    print("Could not convert to int")
````

````python
# Handle file not found
try:
    with open("nonexistent.txt", "r") as f:
        content = f.read()
except FileNotFoundError:
    print("File doesn't exist!")
    content = ""
````

````python
# Multiple exception types + finally
def safe_divide(a, b):
    try:
        result = a / b
    except ZeroDivisionError:
        print("Cannot divide by zero!")
        result = None
    except TypeError:
        print("Invalid types!")
        result = None
    finally:
        print("Division attempted")  # Always runs
    return result

print(safe_divide(10, 2))
print(safe_divide(10, 0))
````
