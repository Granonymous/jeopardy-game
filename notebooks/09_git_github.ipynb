## 1. Why Version Control?

Without version control, you might end up with:
```
project_final.py
project_final_v2.py
project_final_ACTUALLY_FINAL.py
project_final_v2_fixed.py
project_backup_dont_delete.py
```

**Version control gives you:**

| Feature | Benefit |
|---------|--------|
| **History** | See what changed and when |
| **Undo** | Revert any mistake |
| **Branches** | Work on features without breaking main code |
| **Collaboration** | Multiple people can work together |
| **Backup** | Code is stored remotely |

Git is the most popular version control system. GitHub is a website that hosts git repositories.

Think of git as having three "areas":

**Working Directory:** Your actual files on disk. Changes here are "unstaged."

**Staging Area:** Changes you've selected to include in your next commit. Think of it like a shopping cart - you add things to it before "checking out" (committing).

**Repository:** The committed history. This is safe and permanent (unless you really try to delete it).

### Understanding the Three Areas

Why have a staging area? Consider this scenario:

You've been working for an hour. You:
1. Fixed a bug in `game.py`
2. Added a new feature in `cli.py`  
3. Updated some comments in `data.py`

These are three different logical changes! You can use staging to create three separate commits:

```bash
git add game.py
git commit -m "fix: correct score calculation bug"

git add cli.py
git commit -m "feat: add quit command"

git add data.py
git commit -m "docs: clarify data loading comments"
```

This gives you a clean, understandable history!

## 4. Good Commit Messages

Commit messages are for **future you** (and teammates). In 6 months, you'll want to know why a change was made.

### Bad Commit Messages
```
fixed stuff
updates
asdfasdf
wip
.
```

### Good Commit Messages
```
feat: add question filtering by category
fix: correct score calculation for Double Jeopardy
docs: add docstrings to game.py functions
test: add tests for answer normalization
refactor: extract database queries into separate module
```

### Conventional Commit Format
```
type(scope): short description

Longer description if needed.
Explain WHY, not just WHAT.
```

**Types:**
- `feat` - New feature
- `fix` - Bug fix
- `docs` - Documentation
- `test` - Adding/updating tests
- `refactor` - Code change that doesn't add feature or fix bug
- `style` - Formatting, whitespace
- `chore` - Maintenance tasks

## 6. Basic Workflow

Here's the typical workflow you'll use every day:

```bash
# 1. Check what's changed
git status

# 2. Look at the actual changes (optional but helpful)
git diff

# 3. Stage the files you want to commit
git add game.py cli.py
# OR stage everything:
git add .

# 4. Verify what's staged
git diff --staged

# 5. Commit with a good message
git commit -m "feat: add question filtering by category"

# 6. Push to GitHub (if you have a remote)
git push
```

## 7. Common Mistakes and Fixes

Everyone makes these mistakes. Here's how to fix them:

### "I committed something I shouldn't have!"

**If you haven't pushed yet:**
```bash
git reset --soft HEAD~1  # Undo commit, keep changes staged
# Now you can fix and re-commit
```

### "I want to undo my uncommitted changes to a file"
```bash
git checkout -- filename.py  # Discard changes (CAREFUL: can't undo!)
```

### "I committed to the wrong branch!"
```bash
# First, create the correct branch with your commit
git branch correct-branch
# Then reset main to before your commit
git reset --hard HEAD~1
# Switch to correct branch
git checkout correct-branch
```

### "I accidentally staged a secret file!"
```bash
git reset .env  # Unstage the file
# Then add it to .gitignore
```

### "Git says there's a merge conflict!"
Open the file - you'll see markers like:
```
<<<<<<< HEAD
your changes
=======
their changes
>>>>>>> branch-name
```
Edit the file to combine both changes properly, then:
```bash
git add filename.py
git commit -m "resolve merge conflict"
```

## 8. GitHub Basics

GitHub is a website that hosts git repositories. It adds:
- Cloud backup of your code
- Collaboration features (Pull Requests, Issues)
- Project management tools
- Visibility (others can see/use your code)

### Connecting a Local Repo to GitHub

```bash
# 1. Create a repo on GitHub (use their website)
# 2. Connect your local repo:
git remote add origin https://github.com/username/repo.git

# 3. Push your code:
git push -u origin main
```

### Daily Workflow with GitHub

```bash
# Start of day: get any changes from GitHub
git pull

# ... do your work, commit locally ...

# End of day: push your commits
git push
```

## 10. Quick Reference: Git Cheat Sheet

```bash
# Setup
git init                    # Create new repo
git clone <url>             # Copy existing repo

# Daily Use
git status                  # See what's changed
git add <file>              # Stage file
git add .                   # Stage everything
git commit -m "message"     # Commit staged changes
git push                    # Upload to GitHub
git pull                    # Download from GitHub

# Viewing
git log                     # See commit history
git log --oneline           # Compact history
git diff                    # See unstaged changes
git diff --staged           # See staged changes

# Undoing
git reset <file>            # Unstage file
git checkout -- <file>      # Discard changes
git reset --soft HEAD~1     # Undo last commit (keep changes)

# Branches (for later)
git branch                  # List branches
git branch <name>           # Create branch
git checkout <branch>       # Switch branch
git merge <branch>          # Merge branch into current
```
